```python
inventory = {'Espresso': 1.50,
              'Americano':2.00,
              'Caffe latte':2.25,
              'Mocha':2.25,
              'Cappucino':2.50,
              'Macchiato':2.50,
              'White tea': 1.50,
              'Green tea': 2.00,
              'Oolong tea':2.25,
              'Black tea': 2.25,
              'Earl Grey': 2.50,
              'Matcha':2.50,
              'Croissant': 5.50,
              'Cinnamon roll': 4.50,
              'Banana bread': 4.25,
              'Pumpkin bread': 3.75,
              'Sticky bun': 4.25,
              'Apple turnover': 3.75}

basket = []
subtotal = []
frequency_list = {}


#placing the order
def cashier():
    user_answer = input("Would you like to place an order? ")
    while user_answer != "q":
        if user_answer in inventory:
            basket.append(user_answer)
            user_answer = input("Would you like anything else? Type 'q' to end ")
        else:
            user_answer = input("I'm sorry, we don't serve that. Anything else? Type 'q' to end ")

cashier()

for items in basket:
    subtotal.append(inventory[items])
amount = sum(subtotal)

#calculate the tax
tax_rate = float(6/100)
calculated_tax_amount = amount * tax_rate
total = amount + calculated_tax_amount

#print out the receipt
print('-' * 30)
print("Items Order:")
print('-' * 30)

for l in basket:
    if l in frequency_list:
        frequency_list[l] += 1
    else:
        frequency_list[l] = 1

for key in sorted(frequency_list):
    print(f"{key}: {frequency_list[key]}")
print('-' * 30)

print(f"Subtotal: ${amount:.2f}")
print(f"Tax rate(6%): ${calculated_tax_amount:.2f}")
print(f"Total Amount: ${total:.2f}")

print("\nThank you for visiting MC Cafe!")
```
